```python
import datasets
data = datasets.load_dataset("glue", "cola")
data
```

```python
import pandas as pd
import numpy as np
```

```python
train = pd.DataFrame(data['train'])
test = pd.DataFrame(data['test'])
valid = pd.DataFrame(data['validation'])
display(train,test,valid)
```

```python
all_data = pd.concat([train, test, valid])
all_data = all_data.iloc[:,:-1]
all_data
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tk = Tokenizer()
tk.fit_on_texts([all_data['sentence']])
print(tk.word_index)
```
